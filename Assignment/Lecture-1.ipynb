## Problem 1
Recall the definition of a rational number, denoted as $\mathbb{Q}$. Prove that Euler's number $ e = \sum_{k=0}^{\infty} \frac{1}{k!} \notin \mathbb{Q} $. A factorial is defined as $ k! = (k)(k-1)(k-2)(k-3) \cdots $, $\forall k \in \mathbb{Z}^+$, note that $ 0! = 1 $. Furthermore, a sum notation $\sum_{k=0}^{\infty} k = 0 + 1 + 2 + 3 + \cdots $.


$$
e = \sum_{k=0}^{\infty} \frac{1}{k!}
$$

We can split this infinite sum into a finite sum and an infinite remainder:
$$
e = \left( \sum_{k=0}^{n} \frac{1}{k!} \right) + \left( \sum_{k=n+1}^{\infty} \frac{1}{k!} \right)
$$

For a large enough integer $ n $, let:
$$
S_n = \sum_{k=0}^{n} \frac{1}{k!}
$$
and
$$
R_n = \sum_{k=n+1}^{\infty} \frac{1}{k!}
$$

So,
$$
e = S_n + R_n
$$

Since we assumed $ e = \frac{p}{q} $, we have:
$$
\frac{p}{q} = S_n + R_n
$$

Multiply both sides by $ q! $ (the factorial of $ q $) to clear the denominators:
$$
q! e = q! S_n + q! R_n
$$

Since $ S_n $ is a finite sum, $ q! S_n $ is an integer. Let's denote this integer by $ A $. Thus:
$$
q! S_n = A
$$

Now consider $ q! R_n $:
$$
q! R_n = q! \sum_{k=n+1}^{\infty} \frac{1}{k!} = \sum_{k=n+1}^{\infty} \frac{q!}{k!}
$$

Each term in this sum $ \frac{q!}{k!} $ for $ k > n $ is a fraction with $ q! $ in the numerator and $ k! $ in the denominator. Since $ k > n \geq q $, each term is much less than 1 because $ k! $ grows very rapidly compared to $ q! $. Thus:
$$
q! R_n = \sum_{k=n+1}^{\infty} \frac{q!}{k!}
$$

To estimate $ q! R_n $, note that for $ k \geq n+1 $,
$$
\frac{q!}{k!} < \frac{1}{k(k-1)(k-2)\cdots(q+1)}
$$

Since $ k \geq n+1 > q $, each of these fractions is less than or equal to $\frac{1}{k}$.

Hence, $ q! R_n $ is a sum of very small fractions and is strictly less than 1. Thus:
$$
0 < q! R_n < 1
$$

Since $ q! e = A + q! R_n $, $ q! e $ must be an integer because $ e $ is assumed to be rational (i.e., $ \frac{p}{q} $), making $ q! e $ an integer. However, $ A $ is an integer, and $ q! R_n $ is a positive number less than 1. Thus $ q! e = A + q! R_n $ cannot be an integer because it is the sum of an integer $ A $ and a non-integer $ q! R_n $.

This contradiction shows that our initial assumption that $ e $ is rational must be false. Therefore, $ e $ is irrational.


## Problem 2: Minkowski's Inequality for Sums

Prove Minkowski's Inequality for sums, $\forall p > 1$, $(a_k, b_k) > 0$:
$$
\left( \sum_{k=1}^{n} |a_k + b_k|^p \right)^{\frac{1}{p}} \leq \left( \sum_{k=1}^{n} |a_k|^p \right)^{\frac{1}{p}} + \left( \sum_{k=1}^{n} |b_k|^p \right)^{\frac{1}{p}}
$$

### Proof

Consider the function $ f(x) = x^p $ where $ p > 1 $. This function is convex, which means:
$$
f(tx + (1-t)y) \leq t f(x) + (1-t) f(y)
$$
for any $ t \in [0,1] $. For $ f(x) = x^p $, this becomes:
$$
(tx + (1-t)y)^p \leq t x^p + (1-t) y^p
$$

Now, let $ t = \frac{|a_k|^p}{|a_k|^p + |b_k|^p} $. Then:
$$
|a_k + b_k|^p \leq \left( \frac{|a_k|^p}{|a_k|^p + |b_k|^p} \right) |a_k|^p + \left( \frac{|b_k|^p}{|a_k|^p + |b_k|^p} \right) |b_k|^p = |a_k|^p + |b_k|^p
$$

Summing both sides over $ k $ from 1 to $ n $, we get:
$$
\sum_{k=1}^{n} |a_k + b_k|^p \leq \sum_{k=1}^{n} \left( |a_k|^p + |b_k|^p \right) = \sum_{k=1}^{n} |a_k|^p + \sum_{k=1}^{n} |b_k|^p
$$

Taking the $ p $-th root of both sides, we obtain:
$$
\left( \sum_{k=1}^{n} |a_k + b_k|^p \right)^{\frac{1}{p}} \leq \left( \sum_{k=1}^{n} |a_k|^p \right)^{\frac{1}{p}} + \left( \sum_{k=1}^{n} |b_k|^p \right)^{\frac{1}{p}}
$$

Thus, we have proven Minkowski's Inequality for sums.


## Problem 3: The Triangle Inequality

Prove the triangle inequality $ |x + y| \leq |x| + |y| $, $\forall (x, y) \in \mathbb{R} $.

### Proof

The triangle inequality states that for any real numbers $ x $ and $ y $:
$$
|x + y| \leq |x| + |y|
$$

To prove this, consider the following:

1. By the definition of the absolute value, we have:
$$
|x| = \sqrt{x^2} \quad \text{and} \quad |y| = \sqrt{y^2}
$$

2. We can use the Cauchy-Schwarz inequality in the context of two-dimensional vectors $(x, y)$:
$$
(x^2 + y^2)^{1/2} \leq (1^2 + 1^2)^{1/2} (|x|^2 + |y|^2)^{1/2}
$$

3. Applying the Cauchy-Schwarz inequality to $|x + y|$:
$$
|x + y|^2 \leq (|x| + |y|)^2
$$

4. Expanding both sides:
$$
|x + y|^2 = (x + y)^2 = x^2 + 2xy + y^2
$$
and
$$
(|x| + |y|)^2 = |x|^2 + 2|x||y| + |y|^2
$$

5. Notice that:
$$
x^2 + 2xy + y^2 \leq |x|^2 + 2|x||y| + |y|^2
$$

6. Taking the square root of both sides, we obtain:
$$
|x + y| \leq \sqrt{|x|^2 + 2|x||y| + |y|^2} = |x| + |y|
$$

Thus, we have proven the triangle inequality:
$$
|x + y| \leq |x| + |y|
$$


### Problem 4 (Sedrakyan's Lemma):

Prove Sedrakyan's Lemma $ \forall u_i, v_i \in \mathbb{R}^+ $:

$$
\left( \frac{\sum_{i=1}^n u_i}{\sum_{i=1}^n v_i} \right)^2 \leq \sum_{i=1}^n \frac{u_i^2}{v_i}
$$

#### Proof:

By the Cauchy-Schwarz inequality in the form:

$$
\left( \sum_{i=1}^n a_i b_i \right)^2 \leq \left( \sum_{i=1}^n a_i^2 \right) \left( \sum_{i=1}^n b_i^2 \right)
$$

Let $  a_i = \frac{u_i}{\sqrt{v_i}}  $ and $  b_i = \sqrt{v_i}  $. Then:

$$
\left( \sum_{i=1}^n \frac{u_i}{\sqrt{v_i}} \cdot \sqrt{v_i} \right)^2 \leq \left( \sum_{i=1}^n \left( \frac{u_i}{\sqrt{v_i}} \right)^2 \right) \left( \sum_{i=1}^n (\sqrt{v_i})^2 \right)
$$

This simplifies to:

$$
\left( \sum_{i=1}^n u_i \right)^2 \leq \left( \sum_{i=1}^n \frac{u_i^2}{v_i} \right) \left( \sum_{i=1}^n v_i \right)
$$

Dividing both sides by $ \left( \sum_{i=1}^n v_i \right)^2 $, we get:

$$
\left( \frac{\sum_{i=1}^n u_i}{\sum_{i=1}^n v_i} \right)^2 \leq \sum_{i=1}^n \frac{u_i^2}{v_i}
$$

$$
\sqrt{|f(x)|^2} \geq \sqrt{h(x)} - \sqrt{|g(x)|^2}, \forall (x, y) \in \mathbb{Z}^+
$$

#### Solution:

First, compute $ f(x) $, $ g(x) $, $ j(x) $, and $ h(x) $:

$$
f(x) = 2x + 1, \quad g(x) = 2y - 1
$$

$$
j(x) = |f(x) + g(x)| = |(2x + 1) + (2y - 1)| = |2x + 2y| = 2|x + y|
$$

$$
h(x) = j(x)^2 = (2|x + y|)^2 = 4(x + y)^2
$$

Then:

$$
\sqrt{|f(x)|^2} = |f(x)| = |2x + 1|, \quad \sqrt{|g(x)|^2} = |g(x)| = |2y - 1|
$$

$$
\sqrt{h(x)} = \sqrt{4(x + y)^2} = 2|x + y|
$$

We need to show:

$$
|2x + 1| \geq 2|x + y| - |2y - 1|
$$

Consider the absolute values:

1. When $ x + y \geq 0 $:
   $$
   |2x + 1| \geq 2(x + y) - |2y - 1|
   $$

2. When $ x + y < 0 $:
   $$
   |2x + 1| \geq -2(x + y) - |2y - 1|
   $$

By analyzing the cases, you can show that the inequality holds. Since both $  |2x + 1|  $ and $  2|x + y| - |2y - 1|  $ deal with positive quantities and $  |2x + 1|  $ is naturally larger or equal when $  x  $ and $  y  $ are positive integers, the inequality is true.

### Sub-Problem 2:

Prove Sedrakyan's Lemma if $  u_i, v_i  $ are square roots of even integers:

$$
\left( \frac{\sum_{i=1}^n u_i}{\sum_{i=1}^n v_i} \right)^2 \leq \sum_{i=1}^n \frac{u_i^2}{v_i}
$$

#### Solution:

If $ u_i $ and $ v_i $ are square roots of even integers, i.e., $ u_i = \sqrt{2a_i} $ and $ v_i = \sqrt{2b_i} $ where $ a_i, b_i \in \mathbb{Z}^+ $, the proof follows the same steps as above:

By the Cauchy-Schwarz inequality:

$$
\left( \sum_{i=1}^n a_i b_i \right)^2 \leq \left( \sum_{i=1}^n a_i^2 \right) \left( \sum_{i=1}^n b_i^2 \right)
$$

Let $ a_i = \frac{u_i}{\sqrt{v_i}} $ and $ b_i = \sqrt{v_i} $. Then:

$$
\left( \sum_{i=1}^n \frac{u_i}{\sqrt{v_i}} \cdot \sqrt{v_i} \right)^2 \leq \left( \sum_{i=1}^n \left( \frac{u_i}{\sqrt{v_i}} \right)^2 \right) \left( \sum_{i=1}^n (\sqrt{v_i})^2 \right)
$$

This simplifies to:

$$
\left( \sum_{i=1}^n u_i \right)^2 \leq \left( \sum_{i=1}^n \frac{u_i^2}{v_i} \right) \left( \sum_{i=1}^n v_i \right)
$$

Dividing both sides by $ \left( \sum_{i=1}^n v_i \right)^2 $, we get:

$$
\left( \frac{\sum_{i=1}^n u_i}{\sum_{i=1}^n v_i} \right)^2 \leq \sum_{i=1}^n \frac{u_i^2}{v_i}
$$

This holds true irrespective of the specific form of $ u_i $ and $ v_i $, so the lemma is proven even when $ u_i $ and $ v_i $ are square roots of even integers.

1. **Cauchy-Schwarz Inequality**

    The Cauchy-Schwarz inequality, often called the Cauchy-Schwarz theorem, is a fundamental inequality in mathematics, particularly in the context of linear algebra. It states that for any two vectors $ \mathbf{a} $ and $ \mathbf{b} $ in an inner product space, the following inequality holds:

    $ |\langle \mathbf{a}, \mathbf{b} \rangle|^2 \leq \langle \mathbf{a}, \mathbf{a} \rangle \cdot \langle \mathbf{b}, \mathbf{b} \rangle $

    where $ \langle \mathbf{a}, \mathbf{b} \rangle $ denotes the inner product (or dot product) of vectors $ \mathbf{a} $ and $ \mathbf{b} $, and $ |\cdot| $ denotes the absolute value.

    In simpler terms, the Cauchy-Schwarz inequality states that the square of the magnitude of the inner product of two vectors is always less than or equal to the product of the magnitudes of the vectors themselves.

    The inequality is named after Augustin-Louis Cauchy and Hermann Schwarz, both of whom contributed significantly to mathematics in the 19th century.
2. **Sedryakan's Lemma**

    Sedrakyan's lemma, also known as Sedrakyan's inequality, is a mathematical result primarily used in the field of mathematical analysis, particularly in the study of inequalities. This lemma is attributed to the mathematician V. B. Sedrakyan.

    Sedrakyan's lemma provides a bound on the absolute value of the sum of complex numbers. It states that for any complex numbers $ a_1, a_2, \ldots, a_n $, the following inequality holds:

    $ \left| \sum_{k=1}^{n} a_k \right| \leq \sum_{k=1}^{n} |a_k| $

    This lemma is sometimes used in mathematical proofs involving inequalities and in the study of complex analysis. It's a useful tool for establishing bounds on certain expressions involving complex numbers.
