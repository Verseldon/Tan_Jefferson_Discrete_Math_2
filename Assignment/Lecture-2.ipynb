### TASK 1:

*Consider the following experiment:*

1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_i$, $i = 1, 2, \ldots, k$ and remains the same from trial to trial. Notice that $p_1 + p_2 + p_3 + \cdots + p_k = 1$.
4. The trials are independent.
5. The random variables of interest are $Y_1, Y_2, \ldots, Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 + Y_2 + Y_3 + \cdots + Y_k = n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1! y_2! y_3! \ldots y_k!} p_1^{y_1} p_2^{y_2} p_3^{y_3} \ldots p_k^{y_k}$.

**Question 1:**
Show that the expected value $E[Y_i] = \sum_y y_i \cdot P(y) = n \cdot p_i$.

*Solution:*

Let $Y_i$ be the random variable representing the number of trials that fall into cell $i$. We want to find the expected value $E[Y_i]$.

Since $Y_i$ follows a binomial distribution with parameters $n$ and $p_i$, we have:

$$
Y_i \sim \text{Binomial}(n, p_i)
$$

The expected value of a binomially distributed random variable is given by:

$$
E[Y_i] = n \cdot p_i
$$

Thus,

$$
E[Y_i] = \sum_y y_i \cdot P(y) = n \cdot p_i
$$

**Question 2:**

Given the following data:

$$
\begin{array}{|c|c|}
\hline
\text{Age} & \text{Proportion} \\
\hline
18 - 24 & 0.18 \\
25 - 34 & 0.23 \\
35 - 44 & 0.16 \\
45 - 64 & 0.27 \\
65 - 100 & 0.16 \\
\hline
\end{array}
$$

If 500 adults are sampled randomly, find the probability that the sample contains 100 persons between 18 and 24, 200 between 25 and 34, and 200 between 45 and 64. What is the expected value for obtaining a person in the 65 and above?

*Solution:*

We will use the multinomial distribution here. The probability of observing the given sample configuration is:

$$
P(Y_1 = 100, Y_2 = 200, Y_3 = 0, Y_4 = 200, Y_5 = 0) = \frac{500!}{100! \cdot 200! \cdot 0! \cdot 200! \cdot 0!} (0.18)^{100} (0.23)^{200} (0.16)^0 (0.27)^{200} (0.16)^0
$$

Now, calculate the expected number of people aged 65 and above:

Let $Y_5$ be the number of people aged 65 and above. Using the expected value formula for the multinomial distribution:

$$
E[Y_5] = n \cdot p_5
$$

Here, $n = 500$ and $p_5 = 0.16$. So,

$$
E[Y_5] = 500 \cdot 0.16 = 80
$$

So, the expected value for obtaining a person in the 65 and above age group is 80.

1. **Expected Value (Mean):**

   The expected value $ E[Y] $ of a random variable $ Y $ following a Poisson distribution with parameter $ \lambda $ is calculated using the formula:

   $ E[Y] = \sum_{y=0}^{\infty} y \cdot p(y) $

   where $ p(y) $ is the probability mass function of the Poisson distribution.

   For the given distribution $ p(y) = \frac{\lambda^y}{y!} e^{-\lambda} $, we find:

   $ E[Y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} $

   Factoring out $ e^{-\lambda} $ and recognizing the Taylor series expansion of $ e^x $, we rewrite the sum as:

   $ E[Y] = e^{-\lambda} \sum_{y=0}^{\infty} \frac{\lambda^y}{(y-1)!} $

   Applying the property $ \sum_{y=0}^{\infty} \frac{x^y}{y!} = e^x $, we get:

   $ E[Y] = e^{-\lambda} \cdot e^{\lambda} = \lambda $

   Thus, the expected value of this distribution is $ \lambda $.

### Question 2 
2. **Probability of No Seedlings:**

   We are given that the mean density of seedlings is 5 per square yard, so $ \lambda = 5 $. We want to find the probability that none of the 10 sampling regions will contain seedlings.

   Using the Poisson distribution formula, the probability of observing $ y $ events (in this case, seedlings) in a given region is given by:

   $ P(Y = y) = \frac{5^y}{y!} e^{-5} $

   For no seedlings ($ y = 0 $), this simplifies to:

   $ P(Y = 0) = \frac{5^0}{0!} e^{-5} = e^{-5} $

   To find the probability that none of the regions contain seedlings, we calculate:

   $ P(\text{None contain seedlings}) = (e^{-5})^{10} = e^{-50} $

   Therefore, the probability that none of the regions will contain seedlings is approximately $ e^{-50} $.

1. **Poisson Distribution Formula:**
   The Poisson distribution is a probability distribution that describes the number of events occurring in a fixed interval of time or space. It's used when the events are rare and independent of each other. The probability mass function (PMF) of a Poisson distributed random variable $  Y $  with parameter $  \lambda $  is given by:
   $ P(Y = y) = \frac{\lambda^y}{y!} e^{-\lambda} $
   
   Where:
   - $  y $  is the number of events,
   - $  \lambda $  is the average rate (or mean number) of events in the given interval,
   - $  e $  is the base of the natural logarithm, approximately equal to $  2.71828 $ ,
   - $  y! $ denotes the factorial of $  y $, i.e., the product of all positive integers up to $  y $.

2. **Taylor Series Expansion:**
   The Taylor series expansion is a representation of a function as an infinite sum of terms calculated from the values of its derivatives at a single point. It's named after the British mathematician Brook Taylor. The Taylor series expansion of a function $  f(x) $  about the point $  a $  is given by:
   $ f(x) = f(a) + f'(a)(x - a) + \frac{f''(a)}{2!}(x - a)^2 + \frac{f'''(a)}{3!}(x - a)^3 + \dots $
   
   Where:
   - $  f'(a) $ , $  f''(a) $ , $  f'''(a) $ , etc., are the derivatives of $  f(x) $  evaluated at $  x = a $,
   - $  n! $  denotes the factorial of $  n $ .

3. **Binomial Probability Formula**
   The binomial probability formula calculates the probability of getting exactly $ k $ successes in $ n $ independent Bernoulli trials, where each trial has a constant probability $ p $ of success and $ (1-p) $ of failure. In LaTeX, it is represented as:

   $ P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k} $

   Where:
   - $ P(X = k) $ is the probability of getting exactly $ k $ successes,
   - $ \binom{n}{k} $ denotes the binomial coefficient, also known as "n choose k," which represents the number of ways to choose $ k $ successes from $ n $ trials and is calculated as $ \frac{n!}{k! (n - k)!} $,
   - $ p $ is the probability of success in a single trial,
   - $ (1 - p) $ is the probability of failure in a single trial,
   - $ n $ is the number of trials, and
   - $ k $ is the number of successes.
